```python
columns_to_plot = ["VIX", "GTITL30YR", "EONIA","MXJP"]

for column in columns_to_plot:
    plt.figure(figsize=(14, 7))
    ax = sns.lineplot(data=df, x="Data", y=column, label=column, color="blue", linewidth=2)

    for i in range(len(df)):
        if df.loc[i, "Y"] == 1:
            plt.axvspan(df.loc[i, "Data"], df.loc[i, "Data"], color="red", alpha=0.4, label="Crash Period" if i == 0 else "")

    plt.title(f"Market Crash Timelines with {column}", fontsize=18, fontweight="bold")
    plt.xlabel("Date", fontsize=14)
    plt.ylabel(column, fontsize=14)
    plt.legend(fontsize=12, loc="upper left")
    plt.xticks(rotation=45)
    plt.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()
```

```python
X= df.drop(["Y","Data"],axis=1)
y= df["Y"]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import xgboost as xgb
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X = X_scaled
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(),
    "XGBoost": xgb.XGBClassifier()
}

results = {}

for model_name, model in models.items():
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    report = classification_report(y_test, y_pred, output_dict=True)

    results[model_name] = report['accuracy'], report['0']['precision'], report['0']['recall'], report['0']['f1-score'], report['1']['precision'], report['1']['recall'], report['1']['f1-score']

results_df = pd.DataFrame(results, index=["Accuracy", "Precision (0)", "Recall (0)", "F1-Score (0)", "Precision (1)", "Recall (1)", "F1-Score (1)"]).T
results_df.sort_values(by="F1-Score (1)", ascending=False)
```

```python
smt = SMOTE()
X_train_sm, y_train_sm = smt.fit_resample(X_train, y_train)

models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(),
    "XGBoost": xgb.XGBClassifier()
}

results_sm = {}

for model_name, model in models.items():
    model.fit(X_train_sm, y_train_sm)

    y_pred = model.predict(X_test)

    report = classification_report(y_test, y_pred, output_dict=True)

    results_sm[model_name] = report['accuracy'], report['0']['precision'], report['0']['recall'], report['0']['f1-score'], report['1']['precision'], report['1']['recall'], report['1']['f1-score']

results_sm_df = pd.DataFrame(results_sm, index=["Accuracy", "Precision (0)", "Recall (0)", "F1-Score (0)", "Precision (1)", "Recall (1)", "F1-Score (1)"]).T
results_sm_df.sort_values(by="F1-Score (1)", ascending=False)
```
